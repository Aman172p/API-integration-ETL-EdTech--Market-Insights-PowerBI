```python
# changing the data type of the columns('views','likes','comments') into int
def clean_and_impute(df, column_name):
    
    df[column_name] = df[column_name].fillna(0)
    df[column_name] = df[column_name].astype('int')
    mean_val = df[column_name].mean()
    df[column_name] = df[column_name].replace({0: mean_val})
    
    return df
clean_and_impute(data_1,['views','likes','comments'])

data_1["duration"]=data_1["duration"].astype('int')
```
